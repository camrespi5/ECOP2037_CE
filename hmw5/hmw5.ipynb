# Función de costo translogarítmica con tres insumos

1.Considere una función de costo con tres insumos: Labor (pL), Capital (pK) y Fuel (pF ). El nivel de producción es Q.

### 1. Expansión de Taylor de segundo orden

(a) Escriba la expansión de Taylor de segundo orden de lnC(Q; pL; pK; pF ) alrededor de (Q; pL; pK; pF ) = (1; 1; 1; 1), incluyendo términos lineales, cuadráticos y de interacción
en ln Q; ln pL; ln pK; ln pF .

$$
\begin{aligned}
\ln C(Q,p_L,p_K,p_F) \approx\;& 
\alpha_{0} 
+ \alpha_{Q}\,\ln Q 
+ \tfrac{1}{2}\alpha_{QQ}\,(\ln Q)^2 
+ \beta_{L}\,\ln p_L + \beta_{K}\,\ln p_K + \beta_{F}\,\ln p_F \\
&+ \tfrac{1}{2}\gamma_{LL}(\ln p_L)^2 
+ \tfrac{1}{2}\gamma_{KK}(\ln p_K)^2 
+ \tfrac{1}{2}\gamma_{FF}(\ln p_F)^2 \\
&+ \gamma_{LK}\,\ln p_L \ln p_K 
+ \gamma_{LF}\,\ln p_L \ln p_F 
+ \gamma_{KF}\,\ln p_K \ln p_F \\
&+ \theta_{L}\,\ln Q \ln p_L 
+ \theta_{K}\,\ln Q \ln p_K 
+ \theta_{F}\,\ln Q \ln p_F \;.
\end{aligned}
$$


(b) Escriba la interpretación de los parámetros $\alpha_{0}$; $\alpha_{Q}$; $\alpha_{QQ}$; $ \beta_{i} $ ; 
$\gamma_{ij}$ ; $\theta_{i}$.

$\alpha_{0}$ es el nivel de costos base en el punto dado (en logaritmo). $\alpha_{Q}$ es la derivada parcial de lnC respecto a lnQ en el punto. Este se puede interpretar como la elasticidad del costo con respecto a la produccion en el punto. $\alpha_{QQ}$ es la segunda derivada de ln C con respecto a lnQ en el punto. Esto captura como cambia la eslasticidad del costo con Q. $\beta_{i}$ son las derivadas de lnC con respesto a las partipaciones de los insumos. Por el lema de shepard, esto representa las participaciones del costo de cada insumo en el punto base. $\gamma_{ij}$ cuando i=j son las segundas derivadas de lnC con respecto a la participacion de los insumos. Estos miden la cuurvatura de las participaciones compensada por la eslasticidad del insumo. En cambio cuando i es diferente de j, son las segundas derivadas cruzadas, esto representa la sustitucion o complementariedad compensada entre los insumos i y j. Por ultimo, $\theta_{i}$ son las segundas derivadas cruzadas de $∂^2 lnC/∂lnQ ∂lnp_i​$ en el punto. Este se puede interpretar como la sensivilidad del costo a la participacion i varia cuando cambia el nivel de produccion.

### 2. Shares de costo vía Lema de Shephard

Defina los shares de costo $

$$ s_F= \beta_{F} + \gamma_{FF}\ln p_F +  \gamma_{LF}\, \ln p_L + \gamma_{KF}\, \ln p_K + \theta_{F}\,\ln Q $$

(b) Verifique la condición adding–up, es decir, sL + sK + sF = 1.

$$ sL + sK + sF $$
$$
\begin{aligned}
& =(\beta_{L}+ \beta_{K}+ \beta_{F}),  \\
&+ \gamma_{LL}\ln p_L + \gamma_{LK}\ln p_L + \gamma_{LF}\ln p_L,\\
&+ \gamma_{LK}\ln p_K + \gamma_{KK}\ln p_K + \gamma_{KF}\ln p_K,\\

$$

- Por homogeneidad de grado 0: $$\beta_{L}+ \beta_{K}+ \beta_{F}=1$$
- Simetría y restricción de homogeneidad en los γ $$ (\gamma_{LL} + \gamma_{LK} + \gamma_{LF})=0 $$ y lo mismo para K Y F
- Asi mismo, $(\theta_{L} + \theta_{K}, + \theta_{F})=0$

Por ende
$$ sL + sK + sF=1 $$

### 4. Sistema a estimar

(a) Escriba el sistema formado por la ecuación de costo lnC y los shares sL; sK; sF .

**Ecuación de costos:**

$$
\ln C \;=\; \alpha_{0} 
+ \alpha_{Q}\ln Q 
+ \tfrac{1}{2}\alpha_{QQ}(\ln Q)^{2}
+ \sum_{i} \beta_{i}\ln p_{i}
+ \tfrac{1}{2}\sum_{i} \gamma_{ii} (\ln p_{i})^{2}
+ \sum_{i<j} \gamma_{ij}\ln p_{i}\ln p_{j}
+ \sum_{i} \theta_{i}\ln Q \,\ln p_{i}
$$



**Ecuaciones de shares:**

$$
s_{L} \;=\; \beta_{L} 
+ \gamma_{LL}\ln p_{L} 
+ \gamma_{LK}\ln p_{K} 
+ \gamma_{LF}\ln p_{F} 
+ \theta_{L}\ln Q
$$

$$
s_{K} \;=\; \beta_{K} 
+ \gamma_{KK}\ln p_{K} 
+ \gamma_{KL}\ln p_{L} 
+ \gamma_{KF}\ln p_{F} 
+ \theta_{K}\ln Q
$$

$$
s_{F} \;=\; \beta_{F} 
+ \gamma_{FF}\ln p_{F} 
+ \gamma_{FL}\ln p_{L} 
+ \gamma_{FK}\ln p_{K} 
+ \theta_{F}\ln Q
$$


(b) Explique por qué, al estimar, se elimina una ecuación de share (por ejemplo, la de Fuel),
y muestre cómo la función de costo se puede reescribir normalizada por el precio de Fuel
(ln pF ).

Los shares de costo cumplen la condición de adding–up es decir, que las tres ecuaciones de shares no son independientes: cualquiera de ellas puede expresarse como combinación lineal de las otras dos. En términos econométricos, existe multicolinealidad perfecta si intentamos estimar las tres al mismo tiempo.
Por esto, se elimina una ecuación y se estiman solo dos shares, junto con la ecuación de costos. La tercera puede recuperarse por la restricción.

Sistema a estimar normalizado


$$
\begin{aligned}
\ln\!\left(\tfrac{C}{p_F}\right) 
&= \alpha_0 
+ \alpha_Q \ln Q 
+ \tfrac{1}{2}\alpha_{QQ} (\ln Q)^2 \\[6pt]
&\quad + \beta_L \ln\!\left(\tfrac{p_L}{p_F}\right) 
+ \beta_K \ln\!\left(\tfrac{p_K}{p_F}\right) \\[6pt]
&\quad + \tfrac{1}{2}\gamma_{LL}\left[\ln\!\left(\tfrac{p_L}{p_F}\right)\right]^2
+ \tfrac{1}{2}\gamma_{KK}\left[\ln\!\left(\tfrac{p_K}{p_F}\right)\right]^2 \\[6pt]
&\quad + \gamma_{LK}\ln\!\left(\tfrac{p_L}{p_F}\right)\ln\!\left(\tfrac{p_K}{p_F}\right) \\[6pt]
&\quad + \theta_L \ln Q \,\ln\!\left(\tfrac{p_L}{p_F}\right)
+ \theta_K \ln Q \,\ln\!\left(\tfrac{p_K}{p_F}\right)
\end{aligned}
$$




$$
s_{L} \;=\; \beta_{L} 
+ \gamma_{LL}\ln\!\left(\tfrac{p_L}{p_F}\right)
+ \gamma_{LK}\ln\!\left(\tfrac{p_K}{p_F}\right)
+ \theta_{L}\ln Q
$$

$$
s_{K} \;=\; \beta_{K} 
+ \gamma_{KK}\ln\!\left(\tfrac{p_K}{p_F}\right)
+ \gamma_{KL}\ln\!\left(\tfrac{p_L}{p_F}\right)

$$

```python
%%stata

gen lnq       = ln(q)
gen lnpl      = ln(pl)
gen lnpk      = ln(pk)
gen lnpf      = ln(pf)

gen lnc_rel   = ln(cost) - lnpf        // ln(C/pF)
gen lnpl_rel  = lnpl - lnpf            // ln(pL/pF)
gen lnpk_rel  = lnpk - lnpf            // ln(pK/pF)


* --- SUR SIN RESTRICCIONES ---
sureg ///
 (lnc_rel = lnq c.lnq#c.lnq ///
            lnpl_rel lnpk_rel ///
            c.lnpl_rel#c.lnpl_rel c.lnpk_rel#c.lnpk_rel ///
            c.lnpl_rel#c.lnpk_rel ///
            c.lnq#c.lnpl_rel c.lnq#c.lnpk_rel) ///
 (sl      = lnpl_rel lnpk_rel lnq) ///
 (sk      = lnpk_rel lnpl_rel lnq)

```

(b) Imponga las restricciones de homogeneidad, simetría y adding–up, y reestime el sistema.

Como el sistema esta normalizado, las restriccines de homogeneidad y adding up estan implicitas

```python
%%stata

constraint drop _all
constraint define 1 [sL]lnpk_rel = [sK]lnpl_rel

sureg ///
 (lnc_rel = lnq c.lnq#c.lnq ///
            lnpl_rel lnpk_rel ///
            c.lnpl_rel#c.lnpl_rel c.lnpk_rel#c.lnpk_rel ///
            c.lnpl_rel#c.lnpk_rel ///
            c.lnq#c.lnpl_rel c.lnq#c.lnpk_rel) ///
 (sl      = lnpl_rel lnpk_rel lnq) ///
 (sk      = lnpk_rel lnpl_rel lnq), constraints(1)
```

(c) Compare ambos resultados y discuta cuál es más consistente con la teoría.

Al comparar ambos modelos, se observa que los coeficientes estimados prácticamente no cambian entre la versión sin restricciones y la versión restringida; sin embargo, el sistema con restricciones es preferible porque garantiza el cumplimiento de las condiciones teóricas de la función de costos translogarítmica (homogeneidad, simetría y adding–up), sin pérdida de ajuste. Por tanto, aunque empíricamente las diferencias sean mínimas, el modelo restringido resulta más consistente con la teoría microeconómica y más eficiente estadísticamente.

### 6. Elasticidades de sustitución 
Las elasticidades de sustitución miden cómo cambia la combinación
de insumos ante cambios relativos en los precios, manteniendo constante el costo
mínimo.

(a) Usando los parámetros estimados en el punto 5 (con las restricciones impuestas) y los
shares promedio de la muestra, calcule numéricamente:

```python
%%stata

quietly summarize sl sk
scalar sLbar=r(mean)
quietly summarize sk
scalar sKbar=r(mean)
scalar sFbar=1-sLbar-sKbar

scalar g_LL=_b[sl:lnpl_rel]
scalar g_LK=_b[sl:lnpk_rel]
scalar g_KK=_b[sk:lnpk_rel]
scalar g_KL=_b[sk:lnpl_rel]
scalar g_LKsym=(g_LK+g_KL)/2

scalar G_LL=g_LL+sLbar^2-sLbar
scalar G_KK=g_KK+sKbar^2-sKbar
scalar G_LK=g_LKsym+sLbar*sKbar

scalar ec_LL=G_LL/sLbar
scalar ec_KK=G_KK/sKbar
scalar ec_LK=G_LK/sLbar
scalar ec_KL=G_LK/sKbar
scalar ec_LF=-(sLbar*ec_LL+sKbar*ec_LK)/sFbar
scalar ec_KF=-(sLbar*ec_KL+sKbar*ec_KK)/sFbar
scalar ec_FL=(sLbar/sFbar)*ec_LF
scalar ec_FK=(sKbar/sFbar)*ec_KF
scalar ec_FF=-(sLbar*ec_FL+sKbar*ec_FK)/sFbar

scalar sA_LL=ec_LL/sLbar
scalar sA_LK=ec_LK/sKbar
scalar sA_LF=ec_LF/sFbar
scalar sA_KL=ec_KL/sLbar
scalar sA_KK=ec_KK/sKbar
scalar sA_KF=ec_KF/sFbar
scalar sA_FL=ec_FL/sLbar
scalar sA_FK=ec_FK/sKbar
scalar sA_FF=ec_FF/sFbar

scalar M_LK=ec_LK-ec_KK
scalar M_LF=ec_LF-ec_FF
scalar M_KL=ec_KL-ec_LL
scalar M_KF=ec_KF-ec_FF
scalar M_FL=ec_FL-ec_LL
scalar M_FK=ec_FK-ec_KK

di "Elasticidades compensadas"
di "ec_LL=" %9.4f ec_LL " ec_LK=" %9.4f ec_LK " ec_LF=" %9.4f ec_LF
di "ec_KL=" %9.4f ec_KL " ec_KK=" %9.4f ec_KK " ec_KF=" %9.4f ec_KF
di "ec_FL=" %9.4f ec_FL " ec_FK=" %9.4f ec_FK " ec_FF=" %9.4f ec_FF

di "Allen-Uzawa"
di "sA_LL=" %9.4f sA_LL " sA_LK=" %9.4f sA_LK " sA_LF=" %9.4f sA_LF
di "sA_KL=" %9.4f sA_KL " sA_KK=" %9.4f sA_KK " sA_KF=" %9.4f sA_KF
di "sA_FL=" %9.4f sA_FL " sA_FK=" %9.4f sA_FK " sA_FF=" %9.4f sA_FF

di "Morishima"
di "M_LK=" %9.4f M_LK " M_LF=" %9.4f M_LF
di "M_KL=" %9.4f M_KL " M_KF=" %9.4f M_KF
di "M_FL=" %9.4f M_FL " M_FK=" %9.4f M_FK
```

(b) Interprete los resultados: ¿los insumos son sustitutos o complementarios?

Los resultados muestran que las elasticidades compensadas propias (εc\_LL = -0.65, εc\_KK = -0.55 y εc\_FF = -0.03) son negativas, lo cual es consistente con la teoría, porque implican que cuando sube el precio de un insumo, la demanda de ese mismo insumo disminuye. En cuanto a las elasticidades cruzadas, se observa que L y K son sustitutos (εc\_LK = εc\_KL = 0.35 > 0), mientras que la relación con Fuel es positiva pero pequeña (por ejemplo εc\_LF = 0.12 y εc\_KF = 0.08), lo que indica una sustitución débil. Los valores de Allen–Uzawa confirman esta interpretación: entre L y K la elasticidad es relativamente alta (σA\_LK = 1.56), mostrando buena capacidad de sustitución, mientras que entre Fuel y los demás insumos los valores son muy bajos (cercanos a 0), reflejando que apenas hay posibilidad de sustitución. Por último, las elasticidades de Morishima también son positivas en todos los casos (por ejemplo M\_LK = 0.9 y M\_KL = 1), lo cual indica que la tecnología está bien comportada, ya que no aparecen relaciones de complementariedad.


### 7. Intervalos de confianza con el Método Delta

(a) Explique cómo el Método Delta permite obtener la varianza aproximada de una función
no lineal de los estimadores (en este caso, las elasticidades).

El Método Delta permite obtener la varianza aproximada de las elasticidades al linealizar la función no lineal que las define alrededor de los estimadores $\hat{\gamma}_{ij}$, usando la matriz de varianza-covarianza del modelo SUR. Esto se logra calculando las derivadas parciales de la elasticidad con respecto a los parámetros, multiplicándolas por la varianza-covarianza de los estimadores, y tomando la raíz cuadrada para estimar el error estándar, que se usa para construir intervalos de confianza.

(b) Calcule intervalos de confianza al 95% para las elasticidades de sustitución estimadas
en el punto 6.

```python
%%stata 
quietly su sl, meanonly
local sL=r(mean)
quietly su sk, meanonly
local sK=r(mean)
local sF=1-`sL'-`sK'

nlcom (ec_LL:([sl]lnpl_rel+`sL'^2-`sL')/`sL')
nlcom (ec_KK:([sk]lnpk_rel+`sK'^2-`sK')/`sK')
nlcom (ec_LK:(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+`sL'*`sK')/`sL')
nlcom (ec_KL:(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+`sL'*`sK')/`sK')
nlcom (ec_LF:(-([sl]lnpl_rel+0.5*([sl]lnpk_rel+[sk]lnpl_rel))+`sL'*`sF')/`sL')
nlcom (ec_KF:(-(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel)+`sK'*`sF')/`sK')
nlcom (ec_FL:(-([sl]lnpl_rel+0.5*([sl]lnpk_rel+[sk]lnpl_rel))+`sL'*`sF')/`sF')
nlcom (ec_FK:(-(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel)+`sK'*`sF')/`sF')
nlcom (ec_FF:([sl]lnpl_rel+([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel+`sF'^2-`sF')/`sF')

nlcom (sA_LK:(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+`sL'*`sK')/(`sL'*`sK'))
nlcom (sA_LF:(-([sl]lnpl_rel+0.5*([sl]lnpk_rel+[sk]lnpl_rel))+`sL'*`sF')/(`sL'*`sF'))
nlcom (sA_KF:(-(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel)+`sK'*`sF')/(`sK'*`sF'))
nlcom (sA_LL:([sl]lnpl_rel+`sL'^2-`sL')/`sL'^2)
nlcom (sA_KK:([sk]lnpk_rel+`sK'^2-`sK')/`sK'^2)
nlcom (sA_FF:([sl]lnpl_rel+([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel+`sF'^2-`sF')/`sF'^2)

nlcom (M_LK:(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+`sL'*`sK')/`sL'-([sk]lnpk_rel+`sK'^2-`sK')/`sK')
nlcom (M_KL:(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+`sL'*`sK')/`sK'-([sl]lnpl_rel+`sL'^2-`sL')/`sL')
nlcom (M_LF:(-([sl]lnpl_rel+0.5*([sl]lnpk_rel+[sk]lnpl_rel))+`sL'*`sF')/`sL'-([sl]lnpl_rel+([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel+`sF'^2-`sF')/`sF')
nlcom (M_KF:(-(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel)+`sK'*`sF')/`sK'-([sl]lnpl_rel+([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel+`sF'^2-`sF')/`sF')
nlcom (M_FL:(-([sl]lnpl_rel+0.5*([sl]lnpk_rel+[sk]lnpl_rel))+`sL'*`sF')/`sF'-([sl]lnpl_rel+`sL'^2-`sL')/`sL')
nlcom (M_FK:(-(0.5*([sl]lnpk_rel+[sk]lnpl_rel)+[sk]lnpk_rel)+`sK'*`sF')/`sF'-([sk]lnpk_rel+`sK'^2-`sK')/`sK')
```

(c) Interprete: ¿qué significa que el intervalo incluya valores negativos o positivos?

En los IC del 95%: si el intervalo de la elasticidad compensada queda por encima de 0 completamente concluyo sustitución entre i y j. Si queda por debajo de 0 completamente concluyo complementariedad. si incluye 0, el signo no es estadísticamente distinto de cero(relación débil). La misma lectura aplica para Allen usawa si el intervalo >0 indica mayor facilidad de sustitución. Si incluye 0, no puedo afirmar sustitución. Para Morishima si el IC es estrictamente positivo hay sustitución direccional i y j; si incluye 0, el cambio en $x_i/x_j$ no es estadísticamente distinto de cero. Si aparecieran valores negativos, lo tomo como ruido muestral (no como evidencia económica). En las  $\varepsilon^c_{ii}$ espero valores negativos. Si el IC cruza 0, hay poca precisión de la data.
